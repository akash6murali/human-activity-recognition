```python
 def clean_annotation(df):

    parts = df['annotation'].str.split(';')


    df['MET'] = (
        parts.str[-1]
        .str.replace('MET', '', regex=False)
        .astype(float)
    )

    df['code'] = (
        parts.str[-2]
        .str.extract(r'(\d+)')
        .astype(float)
    )

    df['activity'] = parts.str[-3]


    sleeping_mask = df['annotation'].str.contains('sleeping', case=False, na=False)
    df.loc[sleeping_mask, 'activity'] = 'sleeping'
    df.loc[sleeping_mask, 'code'] = 7030
    df.loc[sleeping_mask, 'MET'] = 0.95

    return df
```

```python
def mean_axis_diff(df):
  result=df.groupby('activity')[['x','y','z']].diff().abs().groupby(df['activity']).mean().rename(columns=lambda c:f'mean_diff_{c}')
  return result.reset_index()
```

```python
var_df_p1=mean_axis_diff(p1_sam)
var_df_p1.head(15)
```

```python
var_df_p1_melt=pd.melt(
    var_df_p1,
    id_vars=['activity'],
    value_vars=['mean_diff_x','mean_diff_y','mean_diff_z'],
    var_name='axis',
    value_name='mean_diff'

)

plt.figure(figsize=(15,10))
ax=sns.barplot(data=var_df_p1_melt,x='mean_diff',y='activity',hue='axis')
plt.title('Mean diff per axis')
plt.xlabel('activity')
plt.ylabel('Mean Diff')
for container in ax.containers:
  ax.bar_label(container,fmt='%.3f', label_type='edge', fontsize=9, padding=3)
plt.legend(title='axis')
plt.show()
```

```python
var_df_p13=mean_axis_diff(p13_sam)
var_df_p13.head(15)
```

```python
var_df_p13_melt=pd.melt(
    var_df_p13,
    id_vars=['activity'],
    value_vars=['mean_diff_x','mean_diff_y','mean_diff_z'],
    var_name='axis',
    value_name='mean_diff'

)

plt.figure(figsize=(15,10))
ax=sns.barplot(data=var_df_p13_melt,x='mean_diff',y='activity',hue='axis')
plt.title('Mean diff per axis')
plt.xlabel('activity')
plt.ylabel('Mean Diff')
for container in ax.containers:
  ax.bar_label(container,fmt='%.3f', label_type='edge', fontsize=9, padding=3)
plt.legend(title='axis')
plt.show()
```

```python
p1_sam['next_activity'] = p1_sam['activity'].shift(-1)
transitions = p1_sam.loc[p1_sam['activity'] != p1_sam['next_activity'], ['activity', 'next_activity']].dropna()

transition_counts = pd.crosstab(transitions['activity'], transitions['next_activity'])
transition_matrix = transition_counts.div(transition_counts.sum(axis=1), axis=0)
sns.heatmap(transition_matrix, annot=True, cmap='Blues', fmt='.2f')
plt.title('Activity Transition Matrix (P1)')
plt.show()
```
