# MET Mapping for CAPTURE-24 HAR Project

This notebook defines the MET (Metabolic Equivalent of Task) values for
each activity label in the CAPTURE-24 dataset.

MET values are taken from the 2011 Compendium of Physical Activities 
(Ainsworth et al.) and are used for MET-based evaluation of 
energy expenditure in free-living HAR models.


```python
# MET values based on the 2011 Compendium of Physical Activities

MET_MAP = {
    "sleep": 0.9,             
    "sitting": 1.3,           
    "standing": 1.5,          
    "vehicle": 1.3,           
    "walking": 3.3,           
    "bicycling": 8.0,         
    "household-chores": 3.0,  
    "manual-work": 4.0,       
    "sports": 7.0,            
    "mixed-activity": 3.0     
}

def label_to_met(label: str) -> float:
    return MET_MAP[label]
```

```python
for k,v in MET_MAP.items():
    print(k, "→", v, "MET")
```
